```python
#Splitting data into training and testing subsets
x = np.array(df[["kernel_length", "kernel_width", 
                   "groove_length", "compactness", "asymmetry_coefficient"]])
y = np.array(df["species"])

from sklearn.model_selection import train_test_split
xtrain, xtest, ytrain, ytest = train_test_split(x, y, 
                                                test_size=0.10, 
                                                random_state=42)
```

```python
# training a machine learning model
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(xtrain, ytrain)
print(model.score(xtest, ytest))
```

```python
# prediction
# features = [[kernel_length, "kernel_width", "groove_length", "compactness", "asymmetry_coefficient"]]
features = np.array([[5.137, 2.981, 4.870, 0.8783, 3.631]])
print(model.predict(features))
```
